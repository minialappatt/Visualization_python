Numerical Variables:

    These variables represent quantities and can take on numerical values.
    Examples: age, player ratings (Overall, Potential), height, weight, etc.
    Visualizations for numerical variables-histograms, box plots, scatter plots, and line plots.

Categorical Variables:

    These variables represent categories or groups and do not have an inherent order.
    Examples:nationality, club, preferred foot, position, etc.
    Visualizations for categorical variables: bar charts, pie charts, and count plots.
Ordinal Variables:

    These variables represent categories with a natural order or ranking.
    Examples:jersey number, skill moves, international reputation, etc.
    Visualizations for ordinal variables:ordered bar charts, box plots, and histograms.

### Fill the missing value for the continous variables with Mean(average) for proper data visualization.

```python
# Count the number of missing values in each column
missing_values_count = df.isnull().sum()
print("Number of missing values for each column:")
print(missing_values_count)
```

```python
# Total number of missing values in the entire DataFrame
total_missing_values = missing_values_count.sum()

print("Total number of missing values in the entire DataFrame:", total_missing_values)
```

```python
# Identify continuous variables
continuous_variables = df.select_dtypes(include=['number']).columns.tolist()

# Fill missing values with mean
for column in continuous_variables:
    mean_value = df[column].mean()
    df[column].fillna(mean_value, inplace=True)

# Verify that there are no missing values left
print(df.isnull().sum())
```

### Preprocess height - convert data in format xx'xx to xx.xx Remove "nan" with Mode and convert the column to numerical.


```python
print(df['Height'])
```

```python
# Step 1: Convert height data from 'xx'xx' format to 'xx.xx'
def convert_height(height):
    if pd.isnull(height):
        return height
    feet, inches = map(int, height.split("'"))
    return feet + inches / 12

df['Height'] = df['Height'].apply(convert_height)

# Step 2: Remove "nan" values with Mode
height_mode = df['Height'].mode()[0]
df['Height'].fillna(height_mode, inplace=True)

# Step 3: Convert the column to numerical
df['Height'] = pd.to_numeric(df['Height'])

# Print the preprocessed DataFrame
print(df.head())



```
