```python
Q1 = data['sepal_length'].quantile(0.25)
print(Q1)
```

```python
Q3 = data['sepal_length'].quantile(0.75)
print(Q3)
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
data_noout = data[~((data['sepal_length'] < (Q1-1.5*IQR)) | (data['sepal_length'] > (Q3+1.5*IQR)))]
```

```python
data_noout.shape
```

```python
fig,ax = plt.subplots(2,2,figsize=(10,5))
sns.boxplot(data=data_noout,x='species',y='sepal_length',ax=ax[0][0])
sns.boxplot(data=data_noout,x='species',y='petal_length',ax=ax[0][1])
sns.boxplot(data=data_noout,x='species',y='petal_width',ax=ax[1][0])
sns.boxplot(data=data_noout,x='species',y='sepal_width',ax=ax[1][1])
plt.tight_layout()
```

```python
%reset -f
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("https://raw.githubusercontent.com/AIP-codedb/public/main/data/iris.csv")
data_nonull = data.dropna()
data_noout_z = data_nonull[np.abs(stats.zscore(data_nonull.sepal_length))<=3]
```

```python
data_nonull.isna().sum()
```

```python
fig,ax = plt.subplots(2,2,figsize=(10,5))
sns.boxplot(data=data_noout_z ,x='species',y='sepal_length',ax=ax[0][0])
sns.boxplot(data=data_noout_z ,x='species',y='petal_length',ax=ax[0][1])
sns.boxplot(data=data_noout_z ,x='species',y='petal_width',ax=ax[1][0])
sns.boxplot(data=data_noout_z ,x='species',y='sepal_width',ax=ax[1][1])
plt.tight_layout()
```
