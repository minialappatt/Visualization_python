Numerical Variables:

Categorical Variables:

### Fill the missing value for the continous variables with Mean(average) for proper data visualization.

```python
# Count the number of missing values in each column
missing_values_count = df.isnull().sum()
print("Number of missing values for each column:")
print(missing_values_count)
```

```python
# Total number of missing values in the entire DataFrame
total_missing_values = missing_values_count.sum()

print("Total number of missing values in the entire DataFrame:", total_missing_values)
```

```python
# Identify continuous variables
continuous_variables = df.select_dtypes(include=['number']).columns.tolist()

# Fill missing values with mean
for column in continuous_variables:
    mean_value = df[column].mean()
    df[column].fillna(mean_value, inplace=True)

# Verify that there are no missing values left
print(df.isnull().sum())
```

### Preprocess height - convert data in format xx'xx to xx.xx Remove "nan" with Mode and convert the column to numerical.


```python
print(df['Height'])
```

```python
# Step 1: Convert height data from 'xx'xx' format to 'xx.xx'
def convert_height(height):
    if pd.isnull(height):
        return height
    feet, inches = map(int, height.split("'"))
    return feet + inches / 12

df['Height'] = df['Height'].apply(convert_height)

# Step 2: Remove "nan" values with Mode
height_mode = df['Height'].mode()[0]
df['Height'].fillna(height_mode, inplace=True)

# Step 3: Convert the column to numerical
df['Height'] = pd.to_numeric(df['Height'])

# Print the preprocessed DataFrame
print(df.head())



```

```python
print(df['Height'])
```

### Preprocess weight - convert data in format xxlbs to xx Remove "nan" with Mode and convert the column to numerical.

```python
print(df['Weight'])
```

```python
# Step 1: Remove "lbs" suffix from weight values and replace "nan" values with empty string
df['Weight'] = df['Weight'].astype(str).str.replace('lbs', '').replace('nan', '')

# Step 2: Remove empty strings with Mode
weight_mode = df['Weight'].mode()[0]
df['Weight'].replace('', weight_mode, inplace=True)

# Step 3: Convert the column to numerical
df['Weight'] = pd.to_numeric(df['Weight'])

# Print the preprocessed DataFrame
print(df.head())
```

```python
print(df['Weight'])
```

### Do Univariate anlaysis for outliers detection for height and weight. 

```python
# Univariate analysis for Height
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Height'])
plt.title('Box Plot of Height')
plt.xlabel('Height')
plt.show()
```

```python
# Univariate analysis for Weight
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Weight'])
plt.title('Box Plot of Weight')
plt.xlabel('Weight')
plt.show()

```

```python
import numpy as np

# Calculate z-scores for height and weight
z_scores_height = (df['Height'] - df['Height'].mean()) / df['Height'].std()
z_scores_weight = (df['Weight'] - df['Weight'].mean()) / df['Weight'].std()

# Identify outliers using a threshold (e.g., 3 or -3)
height_outliers = df[np.abs(z_scores_height) > 3]
weight_outliers = df[np.abs(z_scores_weight) > 3]

print("Height outliers:")
print(height_outliers)
print("\nWeight outliers:")
print(weight_outliers)
```

```python
# Calculate IQR for height and weight
Q1_height = df['Height'].quantile(0.25)
Q3_height = df['Height'].quantile(0.75)
IQR_height = Q3_height - Q1_height

Q1_weight = df['Weight'].quantile(0.25)
Q3_weight = df['Weight'].quantile(0.75)
IQR_weight = Q3_weight - Q1_weight

# Identify outliers using IQR
height_outliers = df[(df['Height'] < Q1_height - 1.5 * IQR_height) | (df['Height'] > Q3_height + 1.5 * IQR_height)]
weight_outliers = df[(df['Weight'] < Q1_weight - 1.5 * IQR_weight) | (df['Weight'] > Q3_weight + 1.5 * IQR_weight)]

print("Height outliers:")
print(height_outliers)
print("\nWeight outliers:")
print(weight_outliers)
```

So here we can see there are outliers which need to be removed.  What is shown in the visual is proven with the values.  

```python
# Create subplots for Height and Weight histograms
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot histogram for Height
axs[0].hist(df['Height'], bins=20, color='skyblue', edgecolor='black')
axs[0].set_title('Height Distribution')
axs[0].set_xlabel('Height')
axs[0].set_ylabel('Frequency')

# Plot histogram for Weight
axs[1].hist(df['Weight'], bins=20, color='lightblue', edgecolor='black')
axs[1].set_title('Weight Distribution')
axs[1].set_xlabel('Weight')
axs[1].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
