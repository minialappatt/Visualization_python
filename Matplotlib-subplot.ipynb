```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(start=0,stop=10,num=50)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,label = 'sin')
plt.plot(x,z,label= 'cos')
plt.xlabel('x',fontsize=15,color='blue')
plt.ylabel('sinx',fontsize=15,color='red')
plt.title('Sine Curve')
plt.legend()
```

```python
fig,ax = plt.subplots(2,1)
ax[0].plot(x,y,color='red')
ax[0].set_title('Sine Curve')
ax[1].plot(x,z)
ax[1].set_title('Cosine Curve')
plt.tight_layout()
plt.show()
```
